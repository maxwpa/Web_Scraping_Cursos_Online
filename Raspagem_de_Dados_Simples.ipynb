```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
dfs = []
```

```python
for page_number in range(1, 11):  # Exemplo: páginas de 1 a 5
    url = f'https://carranzacursos.com.br/todos-os-cursos/page/{page_number}/'
    response = requests.get(url)
    
    if response.status_code == 200:
        content = response.content
        site = BeautifulSoup(content, 'html.parser')
        cursos_all = site.findAll('div', attrs={'class': 'product-details-wrap'})

        if cursos_all:
            data_list = []
            for curso in cursos_all:
                titulo = curso.find('h2').text.strip()
                link = curso.find('a')['href']
                if titulo.count('-') == 1:
                    cursos, *cargos = map(str.strip, titulo.split('-', 1))
                else:
                    cursos, *cargos = map(str.strip, titulo.split(' -', 1))
                preco_original = float(curso.find_all('bdi')[0].text.split('R$')[1].replace('.', '').replace(',', '.'))
                preco_atual = float(curso.find_all('bdi')[1].text.split('R$')[1].replace('.', '').replace(',', '.'))
                data = {'Cursos': cursos, 'Cargos': ' '.join(cargos), 'Preço Original': preco_original, 'Preço Atual': preco_atual, 'Link': link}
                data_list.append(data)
                df = pd.DataFrame(data_list)
            dfs.append(df)
```

```python
final_df = pd.concat(dfs, ignore_index=True)
```

```python
final_df.to_csv('Tabela_Cursos.csv', index=False)
```
